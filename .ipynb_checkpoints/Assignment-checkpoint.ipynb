# write a program to calculate the electricity bill (accept number of units from users) according to the following criteria:
unit               Price
first 100 unit - no charge
next 100 units - Rs 5 / unit
after 200 units - Rs 10 / unit

```python
unit = int (input ("enter a unit : "))
if unit <= 100:
    total_bill = 0
elif unit <= 200:
    total_bill = (unit -100) * 5
elif unit > 200:
    total_bill = 500 + (unit - 200) *10
print (total_bill)
```

# Chained conditionals : if , elif , else statements in python

```python
percentage = int (input ("enter a percentage : "))
if percentage < 45:
    print ("the student got failed")
elif percentage >= 45 and percentage < 60:
    print ("the student passed at second division")
elif percentage >= 60 and percentage < 80:
    print ("the student passed at first division")
else:
    print(" the student secured distinction")
```

# BMI Calculator :
the body mass index is calculated by dividing an individuals weight in kilograms by their height in square meters
input :

weight = kg
height = cm

BMI categories:
under weight = < 18.5
normal weight = 18.5 - 25
over weight = 25 - 29.90
obesity = > 30

BMI = kg/m2 where kg is a persons weight & m2 is height in square meter

```python
weight = "80 kg"
height = "177 cm"
```

```python
weight = 80 # kg
height = 177 / 100 # meter
BMI = (weight) / (height * height)
print (round (BMI, 2))
```

```python
BMI = 25.54
if BMI <= 18.5:
    print ("you are under weight")
elif BMI > 18.5 and BMI <= 25:
    print ("you are in normal weight")
elif BMI > 25 and BMI <= 29.90:
    print ("you are over weight")
else:
    print ("you are in obesity")
```
