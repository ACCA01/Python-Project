````python
nums = [8, 3, 4, 5]
sq = []
for item in nums:
    sq.append (item ** 2)
sq
````

````python
nums = [8, 3, 4, 5]
sq = [item ** 2 for item in nums]
sq
````

````python
# Extract positive numbers only
````

````python
num = [1, 1, -5, -1, 3] # iterables
````

````python
pos= [] # normal
for item in num: # for item in iterables
    if item > 0: # conditional expression either true or false
        pos.append(item) # (item) is expression
````

````python
pos
````

````python
# Using comprehension 
# expression = item
# iterable = num
# if true/ false of item is conditional expression
````

````python
pos = [item for item in num if item > 0] # comprehension- expression of item in num
pos
````

````python
x = range (-10 , 11)
````

````python
output = []
for item in x:
    if item <= 0:
        output.append(0)
    else:
        output.append(item)
````

````python
output
````

````python
output = [0 if item <= 0 else item for item in x]
output
````
